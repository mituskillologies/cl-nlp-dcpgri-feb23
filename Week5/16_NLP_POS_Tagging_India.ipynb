```python
import nltk
nltk.download('indian')
```

```python
from nltk.corpus import indian
from nltk.tag import tnt
```

```python
indian.fileids()
```

```python
print('Number of sentences:')
for sent in indian.fileids():
    print(sent)
    print(len(indian.sents(sent)))
```

```python
sents = indian.sents('marathi.pos')
print(sents)
```

```python
pos = indian.tagged_sents('marathi.pos')
for sent in pos:
    print(sent[0][0], sent[0][1])
```

```python
pos
```

```python
tagger = tnt.TnT()
```

```python
tagger.train(pos)
```

```python
sent = 'श्री. प्रभाकर यांनी शिक्षण सुरु केले.'
```

```python
from nltk.tokenize import word_tokenize
words = word_tokenize(sent)
words
```

```python
tagger.tag(words)
```
