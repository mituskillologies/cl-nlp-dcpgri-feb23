```python
import urllib
```

```python
url = urllib.request.urlopen('https://en.wikipedia.org/wiki/Venus')
data = url.read()
```

```python
print(data)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data,'lxml')
```

```python
text = soup.get_text()
```

```python
print(text)
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(text)
```

```python
tokens;
```

```python
from nltk import FreqDist
```

```python
freq = FreqDist(tokens)
```

```python
freq.plot(20, cumulative=False)
```

```python
# Remove the punctuation symbols
```

```python
clean_tokens = []
for token in tokens:
    if token.isalpha() or token.isdigit():
        clean_tokens.append(token)
```

```python
clean_tokens;
```

```python
freq = FreqDist(clean_tokens)
freq.plot(20, cumulative=False)
```

### Stopwords removal

```python
from nltk.corpus import stopwords
```

```python
swords = stopwords.words('english')
```

```python
swords;
```

```python
clean_tokens1 = []
for token in clean_tokens:
    if token.lower() not in swords:
        clean_tokens1.append(token.lower())
```

```python
freq = FreqDist(clean_tokens1)
freq.plot(20, cumulative=False)
```
